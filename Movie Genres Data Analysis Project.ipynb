```python
movies.info()
```

```python
movies.head()
```

```python
movies['profit']=movies['revenue']-movies['budget']
```

```python
movies_genre= movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','release_year','vote_count','vote_average','profit']]
```

```python
movies_genre.head()
```

```python
from pandas import Series, DataFrame
```

```python
split= movies_genre['genres'].str.split('|').apply(Series,1).stack()
split
```

```python
split.index=split.index.droplevel(-1)
split.name='genres_split'
del movies_genre['genres']

movies_genre= movies_genre.join(split)
```

```python
movies_genre
```

```python
#Research Questions (Q)

#1.Which genres are the most common(no. of movies made)?
```

```python
movies_common= movies_genre.groupby('genres_split')['original_title'].count().sort_values(ascending=False)
movies_common
```

```python
sns.countplot(data=movies_genre ,x='genres_split',order = movies_genre['genres_split'].value_counts().index,color='c')
plt.xticks(rotation=90)
plt.show()
```

```python
genres_count= pd.DataFrame(movies_genre.groupby('genres_split').original_title.count()).sort_values(by='original_title',ascending=True)
genres_count
```

```python
genres_count['original_title'].plot.pie(title='Movies per Genre in %',autopct='%1.1f%%', figsize=(10,10))
plt.show()
```

```python
genres_count['original_title'].plot.barh(title= 'Movies Per Genre',color='Darkblue',figsize=(10,9))
plt.show()
```

```python
#2.a.Which genres have the high avg. budget and revenue?
```

```python
movies_avg= movies_genre.groupby('genres_split').mean()
movies_avg
```

```python
movies_avg.sort_values('budget',ascending= True,inplace=True)
```

```python
movies_avg[['budget','revenue']].plot.barh(title= 'Budget and Revenue by Genre',color=('Darkblue','LightBlue'),figsize=(10,9))
plt.show()
```

```python
#2.b.Which genres have the high avg. profit?
```

```python
movies_avg.sort_values('profit',ascending= True,inplace=True)
```

```python
movies_avg['profit'].plot.barh(title= 'Profit by Genre',color='LightBlue',figsize=(10,9))
plt.show()
```

```python
#3.Which genres have the highest avg. popularity?
```

```python
movies_avg.sort_values('popularity',ascending= True,inplace=True)
```

```python
movies_avg['popularity'].plot.barh(title= 'Popularity by Genre',color='Purple',figsize=(10,9))
plt.show()
```

```python
#4.Which genres have the highest number of movies with an voting avg. >=8?
```

```python
movies.head()
movies.drop_duplicates(inplace=True)
movies['profit']=movies['revenue']-movies['budget']
movies_main= movies[['popularity','budget','revenue','original_title','runtime','genres','release_date','vote_count','vote_average','profit']]
```

```python
movies_counted= movies_main[movies_main['vote_count']>=50]

movies_counted.corr(method="spearman")
```

```python
#Just another way to do scatter plot
sns.scatterplot(data=movies_counted,x='vote_average',y='revenue)
```

The vote average does not have high correlation to profit and revenue. So high voting average would not really mean high profit and high revenue.

```python
#5.Look at the profit for each Genres by release year.
```

```python
movies_genre.head()
```

```python
movie_year= pd.DataFrame(movies_genre.groupby(['release_year','genres_split'])['profit'].mean())
movie_year
```

```python
final_genre= pd.pivot_table(movie_year,values='profit',index='genres_split',columns='release_year')
final_genre
```

```python
fig, ax = plt.subplots(figsize=(15,10))
sns.heatmap(final_genre,cmap='YlGnBu',linewidths=0.5)
plt.title('Genres by Profit per Year')
plt.show()
```
